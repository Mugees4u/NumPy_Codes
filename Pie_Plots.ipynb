#Customization Pie Charts

- Startangle : This attribute allows to rotate the Pie chart in python counterclockwise around the x-axis by the specified degrees.

- Shadow : This is boolean attribute adds a shadow effect below the rim of the pie. Setting the True can make your chart stand out and give it more 3D appearence.

- wedgeprops : customize the each wedges.

- frames : When set True, draws a frame around a pie chart.

- autopct : it controls how the percentage displayed on the wedges.



```python
import numpy as np
import matplotlib.pyplot as plt

cars = ['Alto', 'Seltos', 'Thar', 'Defenders', 'Skoda', 'Ciaz', 'Audi']
data = [100, 56, 15, 18, 29, 43, 42]

#create explode data
explode = [0.1, 0.0, 0.2, 0.3, 0.0, 0.1, 0.4]

#creating color parameters
colors = ('pink', 'indigo', 'blue', 'Green', 'yellow', 'Red', 'orange')

#Wedge Properties
wp = {'linewidth':1, 'edgecolor': 'green'}

#creating autopc arguments

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

#creating plot
fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                 autopct=lambda pct: func(pct, data),
                                 explode=explode,
                                 labels=cars,
                                 shadow=True,
                                 colors=colors,
                                 startangle=90,
                                 wedgeprops=wp,
                                 textprops=dict(color="black"))

#Adding legend
ax.legend(wedges, cars,
         title="Cars",
         loc="center left",
         bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight='bold')
ax.set_title('Customizing Pie Chart')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

PureGold = ['Jashn-E-Mehfil (Basmati Rice)', 'Shahi Pakwan (Basmati Rice)', 'Sultan (HMT Rice)', 'Maharaj (Sona Masoori)']
data = [340, 510, 1000, 1500]

#create explode data
explode = [0.1, 0.0, 0.2, 0.3]

#creating color parameters
colors = ('Red', 'Blue', 'Green', 'Pink')

#Wedge Properties
wp = {'linewidth':1, 'edgecolor': 'green'}

#creating autopc arguments

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

#creating plot
fig, ax = plt.subplots(figsize=(8, 5))
wedges, texts, autotexts = ax.pie(data,
                                 autopct=lambda pct: func(pct, data),
                                 explode=explode,
                                 labels=PureGold,
                                 shadow=True,
                                 colors=colors,
                                 startangle=90,
                                 wedgeprops=wp,
                                 textprops=dict(color="Black"))

#Adding legend
ax.legend(wedges, PureGold,
         title="PureGold",
         loc="center left",
         bbox_to_anchor=(1.2, 0.4, 0.5, 1))

plt.setp(autotexts, size=8, weight='bold')
ax.set_title('Pure Gold Long Grain Rice')

plt.show()
```

# Labels

The labels to the pie chart with the labels parameters.


Note: The labels parameters must be an array with one label for each wedge.

```python
#Labels
import matplotlib.pyplot as plt
import numpy as np

x = np.array([13, 65, 12, 16])

y = ['Apple', 'Carrot', 'Juice', 'Mango']

plt.pie(x, labels=y)
plt.show()
```

# Start Angle

default start angle is x-axis, but we can change the start angle by specifying a 'startangle' parameter.

The 'startangle' parameter is defined with an angle in degree, default angle is zero

```python
#Example - Start first wedge at 90 deg.

import matplotlib.pyplot as plt
import numpy as np

x  = np.array([10, 26, 45, 36, 98])

y = ['Apple', 'Kiwi', 'Water Apple', 'Pineapple', 'Water Mellon']

plt.pie(x, labels=y, startangle=90)
plt.show()
```

```python
#Example - Start first wedge at 135 deg.

import matplotlib.pyplot as plt
import numpy as np

x  = np.array([17, 26, 45, 36, 98])

y = ['Apple', 'Kiwi', 'Water Apple', 'Pineapple', 'Dates']

plt.pie(x, labels=y, startangle=135)
plt.show()
```

# Explode

Explode is used when you want any wedge stand out. The 'explode' parameter is used.

The 'explode' parameter, if specified, and note 'None', must be an array with one value for each wedge.

Each value represents how for from the center each wedge is displayed.


```python
#Example : One Wedge explode 0.2

import numpy as np
import matplotlib.pyplot as plt

rank = np.array([1, 2, 3, 4])
country = ['India', 'Pakistan', 'Srilanka', 'Afghanistan']

#creating Explode
exp = [.2, 0, 0, 0]

plt.pie(rank, labels=country, startangle=270, explode= exp)
plt.show()
```

```python
#explode .3, .2, .1
import matplotlib.pyplot as plt
import numpy as np

rice = np.array([1, 2, 3])
brands = ['Jashn-E-Mehfil (1121)', 'Shahi Pakwn (1509)', 'Daawat-E-Khas (1401)' ]

expl = [0.3, .1, .2]

plt.pie(rice, labels=brands, startangle=145, explode = expl)

plt.show()
```

# Shadow

Shadow to the pie chart by setting the 'shadows' parameter to 'True'

```python
#Example: shadows to a pie chart

import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3])
sex = ['Male', 'Female', 'Transgenders']
expld = [.3, 0, 0]

plt.pie(x, labels=sex, explode = expld, shadow = True)

plt.show()
```

# Colors

We can set the colors for each wedge with the 'colors' parameter.

The 'colors' parameter if specified must be an array with one value for each wedge.

```python
#Example : specify the color for each wedge

c = [1, 1, 1]
l = ['Stop', 'Go', 'Go-Slow']
clrs = ['r', 'g', 'orange']
plt.pie(c, labels=l, colors=clrs, shadow = True)
plt.show()
```

# Legend
