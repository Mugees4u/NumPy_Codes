# Pyplot:
   It is a submodule and are usually imported under the 'plt' alias

```python
import matplotlib.pyplot as plt
```

```python
#Exmple : Draw a line in a diagram from position (0, 0) to (6, 250)
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)

plt.show()
```

```python
x = np.array([10, 8])
y = np.array([42, 36])

plt.plot(x, y)
plt.show()
```

```python
#Example: Draw a 2 points in the diagram, one at pos. (1, 3) & (8, 10)

x = np.array([1, 8])
y = np.array([3, 10])

plt.plot(x, y, 'o')

plt.show()
```

```python
x = [1, 2, 6, 8]
y = [3, 8, 1, 10]

plt.plot(x, y)
plt.show()
```

# Markers
   we can use the keyword argument 'marker' to emphasize each point with a specified marker

```python
t = [6, 8, 12, 9]

plt.plot(y, marker = '^')
plt.show()
```

# Format Strings 'fmt'
   'fmt' written as syntax:
               
                   marker|line|color

```python
x = [6, 19, 27, 13, 21]
plt.plot(x, '^--g')
plt.show()
```

# Marker Size
      we can use the keyword argument 'markersize' or 'ms' to set the size of markers

```python
x = [25, 48, 32, 64, 31, 29]
plt.plot(x, '^:r', ms=20)
plt.show()
```

# Marker Color
    The keyword argument 'markeredgecolor' or 'mec' to set the edge color of the marker

```python
z = [3, 9, 5, 11]
plt.plot(z, 'o--b', ms=10, mec='r')
plt.show()
```
