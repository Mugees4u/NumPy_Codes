# Dot Product

Dot product is an algebric operations in which two equal length vectors are being multiplied such that is produced a single number.
            Dot product often called inner product.

```python
#Dot Product

import time
import numpy
import array
```

```python
a = array.array('q')
```

```python
for i in range(100000):
    a.append(i)
```

```python
b = array.array('q')
for i in range(100000, 200000):
    b.append(i)
```

```python
#classic dot products of vectors impleentations
tic = time.process_time()
dot = 0;
```

```python
for i in range(len(a)):
    dot += a[i] * b[i]
```

```python
toc = time.process_time()
```

```python
print("dot product ="+ str(dot));
```

```python
print("computation time =" + str(1000*(toc - tic)) + "ms")
```

```python
n_tic = time.process_time()
```

```python
n_dot_product = numpy.dot(a, b)
```

```python
n_toc = time.process_time()
```

```python
print("\n dot produt = "+ str(n_dot_product))
```

```python
print("Computation time = "+ str(1000*(n_toc - n_tic))+ "ms")
```

```python
#Example:-
import numpy as np
def for_loop_multiply_lists(list1, list2):
    product_list = []
    for element1, element2, in zip(tuple1, tuple2):
        product_list.append(element1 * element2)
    return product_list

def multiply_arrays(array1, array2):
    return array1 * array2
```

```python
list1 = list(range(1000))
```

```python
list2 = list(range(1000))
```
