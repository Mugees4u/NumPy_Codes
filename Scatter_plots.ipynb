# Scatter Plots

This plot is used to visualize the relationship between two variables. A Symbol dot is used to show a data pair.

relationship among independent or dependent variables

The 'scatter()' plot function plots one dot for each observation. It needs 2 arrays of the same length.

```python
import numpy as np
import matplotlib.pyplot as plt

Hours_of_Study = [2, 4.5, 5, 5, 6, 3, 10, 9.5, 8]
Exam_Score = [53, 35, 91, 72, 60, 62, 85, 78, 99]

#plot the figure
fig = plt.figure(figsize = (8, 3))

plt.scatter(Hours_of_Study, Exam_Score)
plt.show()
```

```python
#Example: A simple plot

x = np.array([5, 10, 7, 19, 21, 9, 8, 22, 13, 14, 18, 21])

y = np.array([53, 65, 87, 61, 73, 82, 94, 67, 49, 67, 58, 84])

plt.scatter(x, y)
plt.show()
```

```python
#Example : Draw plots on the same figure

#data1
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 17, 26, 23, 54, 89, 98, 25, 89, 18, 21, 53])

plt.scatter(x, y)

#data2
x = np.array([2, 2, 21, 17, 13, 21, 16, 8, 9, 3, 7, 18, 16, 24, 25])
y = np.array([77, 89, 61, 87, 61, 99, 67, 81, 91, 35, 64, 89, 77, 63, 93])

plt.scatter(x, y)

plt.show()
```

# colors

We can set our own color for each scatter plots with the 'color' or 'c' arguments.

bydefault color is 'blue' or 'orange'

```python
#Example - set the own colors

x = [2, 4, 6, 8, 10]
y = [21, 23, 25, 27, 29]

plt.scatter(x,y, color = 'hotpink')

x = [1, 3, 5, 7, 9]
y = [22, 24, 26, 28, 30]

plt.scatter(x, y, color = 'green')

plt.show()
```

# Colors Each dot

We can set different or specific color for each dot by using the 'c' arguments.

But we cannot use here 'color' argument for this

```python
rank = [56, 44, 21, 16, 1]
year = np.array([2020, 2021, 2022, 2023, 2024], dtype=int)

colors = ['hotpink', 'silver', 'black', 'green', 'red']
#fig = plt.figure(figsize=(12,7))
plt.scatter(year, rank, c=colors)

plt.show()
```

# ColorMap
#### A color map just like a list of colors, where each color  has a value that range from 0 to 100 
#### The color map is called 'viridis'.


We can specifiy the colormap with the keyword argument 'cmap' with the value of colormap

```python
#Example - create a color array, and specify the colormap in the scatter plot

x = [14, 26, 29, 12, 27]
y = [5, 8, 9, 4, 8.5]
colors = [0, 18, 54, 78, 100]

plt.scatter(x, y, c=colors, cmap='viridis')
plt.show()
```

```python
#Example - include the actual color bar

x = [14, 26, 29, 12, 27]
y = [5, 8, 9, 4, 8.5]
colors = [0, 18, 54, 78, 100]

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```
