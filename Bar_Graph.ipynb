# Multiple bar Plots

```python
barWidth = 0.25
fig = plt.subplots(figsize =(8, 6))

IT = [12, 30, 1, 8, 22]
ECE = [26, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]


br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1, IT, color='r', width=barWidth, edgecolor = 'gray', label='IT')
plt.bar(br2, ECE, color='g', width=barWidth, edgecolor = 'gray', label = 'ECE')
plt.bar(br3, CSE, color='b', width=barWidth, edgecolor = 'gray', label = 'CSE')


plt.xlabel('Branch', fontweight = 'bold', fontsize = 15)
plt.ylabel('Student Passed', fontweight = 'bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IT))], ['2015', '2016', '2017', '2018', '2019']) 

plt.legend()
plt.show()
```

Staked Bar Plots

- it represnt the different groups on top of one another.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5   # groups of bars

boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)
boyStd = (2, 3, 4, 1, 2)
girlStd = (3, 5, 2, 3, 3)

ind = np.arange(N)
width = 0.3


#Plot a subplots

fig = plt.subplots(figsize = (10, 7))

p1 = plt.bar(ind, boys, width, yerr = boyStd)
p2 = plt.bar(ind, girls, width, yerr = girlStd, bottom = boys)


plt.ylabel('Contribution')
plt.title('Contribution by the team')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))
plt.show()
```

```python
#Example:- Simple Stacked bar plot

x = ['A', 'B', 'C', 'D', 'E']
y1 = [10, 20, 30, 40, 23]
y2 = [65, 58, 32, 14, 59]


plt.bar(x, y1, color='b', width=0.5)
plt.bar(x, y2, color='g', width=0.5, bottom = y1)

plt.show()
```

```python
#Example:- Stacked bar chart with more than 2 data


import numpy as np
import matplotlib.pyplot as plt



#create a data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([30, 65, 95, 12])
y3 = np.array([45, 69, 74, 98])
y4 = np.array([65, 78, 86, 54])

plt.bar(x, y1, color='r', width=0.4)
plt.bar(x, y2, color='g', bottom=y1, width=0.4)
plt.bar(x, y3, color='b', bottom=y1+y2, width=0.4)
plt.bar(x, y4, color='y', width=0.4, bottom=y1+y2+y3)


plt.xlabel('Team')
plt.ylabel('Score')
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4", "Round 5"])
plt.title('Socores by Teams in 4 Rounds')

plt.show()
```

```python
#Example: Plot Stacked bar using DataFrames

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#create data

df = pd.DataFrame([['A', 10, 20, 26, 30],
                  ['B', 75, 38, 65, 12],
                  ['C', 78, 45, 25, 11],
                  ['D', 14, 19, 78, 24]],
                 columns=['Team', 'Round 1', 'Round 2', 'Round 3', 'Round 4'])

print(df)


#plot stacked bar

df.plot(x='Team', kind='bar', stacked=True, title='Stacked Bar Graph by dataframes')

plt.show()
```
